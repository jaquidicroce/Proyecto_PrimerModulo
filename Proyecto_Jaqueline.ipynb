## Descripción del Dataset

El dataset incluye las siguientes características:

- **Campaign_ID**: Identificador único de cada campaña.
- **Channel**: Canal de marketing utilizado (e.g., Social Media, Email, TV, News, Radio, Billboard).
- **Product_Category**: Categoría del producto (e.g., Electronics, Fashion, Groceries, Automotive, Health).
- **Region**: Región geográfica en Malasia donde se dirigió la campaña (e.g., Kuala Lumpur, Penang, Malacca, Johor, Sabah, Sarawak).
- **Budget**: Presupuesto asignado en Ringgit malayo.
- **Spend**: Gasto real de la campaña en Ringgit malayo.
- **Impressions**: Número de veces que se vio el anuncio.
- **Clicks**: Número de clics que recibió el anuncio.
- **CTR**: Tasa de clics (Click Through Rate).
- **Conversions**: Número de acciones exitosas (e.g., compras, registros).
- **CR**: Tasa de conversión (Conversion Rate).
- **Revenue**: Ingresos generados por la campaña en Ringgit malayo.
- **ROI**: Retorno de la inversión (Revenue dividido por Spend).
- **Duration**: Duración de la campaña en días.
- **Start_Date**: Fecha de inicio de la campaña.
- **End_Date**: Fecha de finalización de la campaña.
- **Customer_Satisfaction**: Calificación de satisfacción del cliente (1 a 5).
- **Campaign_Effectiveness**: Efectividad de la campaña según el equipo de marketing (1 a 10).
- **Market_Segment**: Segmento de mercado objetivo (Low, Medium, High).

**Nota**: El dataset incluye valores faltantes para simular escenarios del mundo real.

- Utiliza gráficos de caja (boxplots) para identificar outliers en variables numéricas.

  - **Mantenerlos** si son relevantes.
  - **Eliminarlos** si son errores.
  - **Transformarlos** utilizando métodos como logaritmos.

```python
# Outliers = valores atípicos que podrian afectar el análisis
# Gráfica de caja para identificar outliers

plt.figure(figsize=(15,10))
sns.boxplot(data=df, vert=False, orient='h')

# Ajustar el eje x para que se vea mejor la gráfica

plt.xscale('log')
plt.show()
```

```python
# Para los casos de Customer_Satisfaction y Campaign_Effectiveness, eliminaré los outliers porque estas columnas tienen valores limitados.

# Calcular los límites para customer_satisfaction
Q1_cs = df['Customer_Satisfaction'].quantile(0.25)
Q3_cs = df['Customer_Satisfaction'].quantile(0.75)
IQR_cs = Q3_cs - Q1_cs
lower_bound_cs = Q1_cs - 1.5 * IQR_cs
upper_bound_cs = Q3_cs + 1.5 * IQR_cs

# Calcular los límites para campaign_effectiveness
Q1_ce = df['Campaign_Effectiveness'].quantile(0.25)
Q3_ce = df['Campaign_Effectiveness'].quantile(0.75)
IQR_ce = Q3_ce - Q1_ce
lower_bound_ce = Q1_ce - 1.5 * IQR_ce
upper_bound_ce = Q3_ce + 1.5 * IQR_ce

# Filtrar los datos para eliminar los valores fuera del rango en customer_satisfaction y campaign_effectiveness
df_filtered = df[
    (df['Customer_Satisfaction'] >= 1) & (df['Customer_Satisfaction'] <= 5) &
    (df['Campaign_Effectiveness'] >= 1) & (df['Campaign_Effectiveness'] <= 10)
]

# df_filtered ahora contiene los datos sin valores fuera de los rangos especificados
```

```python
# Función para eliminar outliers usando el método IQR
# Eliminamos los outliers de las columnas customer satisfaction y campaign effectivenes porque son valores limitados

def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Eliminar outliers en las columnas customer_satisfaction y campaign_effectiveness
df = remove_outliers(df, 'Customer_Satisfaction')
df = remove_outliers(df, 'Campaign_Effectiveness')
```

```python
# Realizar analisis de clusters para identificar patrones en los datos

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Seleccionar las columnas numéricas para el análisis de clusters
columns = ['Spend', 'Impressions', 'Clicks', 'Conversions', 'Revenue', 'ROI', 'Duration', 'Customer_Satisfaction', 'Campaign_Effectiveness']

# Crear un nuevo DataFrame solo con las columnas seleccionadas
cluster_df = df[columns]

# Escalar los datos para que tengan media 0 y desviación estándar 1
scaler = StandardScaler()
scaled_data = scaler.fit_transform(cluster_df)

# Crear un DataFrame con los datos escalados
scaled_df = pd.DataFrame(scaled_data, columns=columns)

# Realizar el análisis de clusters con 5 clusters
kmeans = KMeans(n_clusters=5, random_state=0)
kmeans.fit(scaled_df)

# Agregar las etiquetas de los clusters al DataFrame original
df['Cluster'] = kmeans.labels_

# Mostrar el conteo de campañas por cluster
df['Cluster'].value_counts()

# Graficar la relación entre las variables Spend y Revenue, coloreando por cluster

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Spend', y='Revenue', hue='Cluster', palette='pastel')
plt.title('Relación entre Spend y Revenue por Cluster')
plt.xlabel('Spend')
plt.ylabel('Revenue')
plt.ticklabel_format(style='plain', axis='y')
plt.show()

# Graficar la relación entre las variables Impressions y Clicks, coloreando por cluster

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Impressions', y='Clicks', hue='Cluster', palette='pastel')
plt.title('Relación entre Impressions y Clicks por Cluster')
plt.xlabel('Impressions')
plt.ylabel('Clicks')
plt.ticklabel_format(style='plain', axis='y')
plt.show()

# Graficar la relación entre las variables Conversions y Revenue, coloreando por cluster

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Conversions', y='Revenue', hue='Cluster', palette='pastel')
plt.title('Relación entre Conversions y Revenue por Cluster')
plt.xlabel('Conversions')
plt.ylabel('Revenue')
plt.ticklabel_format(style='plain', axis='y')
plt.show()

# Graficar la relación entre las variables ROI y Duration, coloreando por cluster

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='ROI', y='Duration', hue='Cluster', palette='pastel')
plt.title('Relación entre ROI y Duration por Cluster')
plt.xlabel('ROI')
plt.ylabel('Duration')
plt.ticklabel_format(style='plain', axis='y')
plt.show()





```
